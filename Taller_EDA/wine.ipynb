```python
#importe la libreria de pandas para imprimir el archivo de vinos en los cuales junte las dos bases de datos en una y de esta manera tener una mejor visualizacion de los datos
import pandas as pd
path= 'winequality-white.csv'
df = pd.read_csv(path)
df
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#En la grafica realiza con lineplot correlacion el acohol y ph de los vinos, se puede evidenciar que entre el 9-10% de alcohol tiene mas ph por ende tendra una mayor alcalinidad
df2 = df.groupby('alcohol').sum()
df2
sns.lineplot(data=df2, x="alcohol", y="pH")
```

```python
#En  la grafica pointplot lo realice con los campos de type y residual sugar, se evidencia que los vinos blancos tienen un mayor residuo de azucar el cual es la cantidad total de azÃºcar que queda en el vino que no ha sido fermentada por las levaduras.
sns.pointplot(data=df, x="type", y="residual sugar")
plt.show()
```

```python
#La tercer grafica realizada con coutplot podemos observar que hay una mayor produccion de vino blanco que rojo.
sns.countplot(x=df["type"])
```

```python
#Como cuarta grafica hice una correlacion entre type y alcohol, se puede analizar que el vino blanco tiene un porcentaje de alcohol mas alto  a comparacion del vino rojo, los valores los obtuvimos realizando la media del acohol
import plotly.express as px
df['mean_white'] = df.query('type == "white"')['alcohol'].mean()
df['mean_red'] = df.query('type == "red"')['alcohol'].mean()
fig = px.bar(df, x='type', y='mean_white')
fig.show()
```

```python
#Como ultima grafica utilice scatterplot relacionando los campos de citric acid y type, donde podemos observar que tienen una cantidad de acido citrico muy parecido entre el vino blanco y rojo, el cual aporta sensaciones de frescura, contribuyen-do al equilibrio gustativo del vino.
sns.scatterplot(data=df, x="citric acid", y="type")
```
