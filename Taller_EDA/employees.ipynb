```python
#importe la libreria de pandas para imprimir el archivo de employees y de esta manera tener una mejor visualizacion de los datos
import pandas as pd
path= 'employees.csv'
df = pd.read_csv(path)
df
```

```python
#Por medio de esta funcion podemos saber el valor medio, maximo, minimo, cuartiles que tienen los datos de employees, solo se puede hacer con valores de tipo int o float
df.describe()
```

```python
df.info()
```

```python
#De esta manera podemos saber que tipo de variables en cada columna del archivo de employees
print(df.dtypes)
```

```python

# En esta linea de codigo reemplace los espacios NaN de las columnas: genero, First Name, Salary y Bonus, estas dos ultimas con la media
df["Gender"].fillna("No Gender", inplace = True)
df["First Name"].fillna("No name", inplace = True)
df['Salary'].fillna(int(df['Salary'].mean()), inplace=True) 
df['Bonus %'].fillna(float(df['Bonus %'].mean()), inplace=True)  
df
```

```python
#Realice el cambio de NaN por medio de la moda en los campos de Senior Management
import numpy as np
mode = df["Senior Management"].mode().values[0]
df["Senior Management"]= df["Senior Management"].replace(np.nan, mode)
df
```

```python
#Realice cambios en los espacio NaN por medio de la moda en los campos de Team
import numpy as np
mode = df["Team"].mode().values[0]
df["Team"]= df["Team"].replace(np.nan, mode)
df
```

```python
df[['Salary','Bonus %']].corr()
```

```python
#En esta linea de codigo se puede evidenciar la agrupacion y la suma principalmente de genero y salario
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 

df3 = df.groupby('Gender').sum().reset_index()
df3
```

```python
#En la grafica realiza con lineplot podemos observar que en los años 1995 y 2010 fueron los años donde los empleados obtuvieron mejores salarios 
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
df['Year'] = df['Start Date'].dt.year

df2 = df.groupby('Year').sum()

sns.lineplot(data=df2, x="Year", y="Salary")
```

```python
#En  la grafica pointplot lo realice con los campos de genero y salario, se evidencia que las mujeres y los hombres tienen un salario muy parecido como empleados de esta empresa.
sns.pointplot(data=df3, x="Gender", y="Salary")
plt.show()
```

```python
#La tercer grafica realizada con coutplot en la cual utilice los datos de genero se puede observar que hay mas mujeres trabajando en la empresa que hombres aunque no es una diferencia marcada.
sns.countplot(x=df["Gender"])
```

```python
#Como cuarta grafica hice una correlacion entre senior management y salary en la cual se evidencia que las personas que si son senior management tienen un mejor salario a comparacion de las personas que no lo son.
import plotly.express as px
fig = px.bar(df, x='Senior Management', y='Salary')
fig.show()
```

```python
#Como ultima grafica utilice scatterplot relacionando Bonus % y Gender donde se observa que no hay gran diferencia entre los generos al ganar los bonos, se ven un poco parejos.
sns.scatterplot(data=df, x="Bonus %", y="Gender")
```
